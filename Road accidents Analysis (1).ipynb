## 10) Number of Accidents, people KILLED, INJURED on SINGLE LANE per 1L population

```python
def calculate_per_100k(total_value, population):
    return (total_value / population) * 100000

def calculate_single_lane(data):
    accidents_single_lane = data['Single Lane - Accident - 2014']
    killed_single_lane = data['Single Lane - Killed - 2014']
    injured_single_lane = data['Single Lane - Injured - 2014']
    population = data['Population']

    accidents_per_100k = calculate_per_100k(accidents_single_lane, population)
    killed_per_100k = calculate_per_100k(killed_single_lane, population)
    injured_per_100k = calculate_per_100k(injured_single_lane, population)

    return accidents_per_100k, killed_per_100k, injured_per_100k

state_data ={ 
    'Single Lane - Accident - 2014': 8634,
    'Single Lane - Killed - 2014': 2958,
    'Single Lane - Injured - 2014': 13541,
    'Population': 52410653
}


accidents_100k, killed_100k, injured_100k = calculate_single_lane(state_data)
print(f'Accidents on Single Lane per 100,000 population: {accidents_100k:.2f}')
print(f'People Killed on Single Lane per 100,000 population: {killed_100k:.2f}')
print(f'People Injured on Single Lane per 100,000 population: {injured_100k:.2f}')
```
